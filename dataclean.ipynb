```python
import pandas as pd

# Load your dataset (assuming it's in Excel)
df = pd.read_excel("Newcleandata.xlsx")

# Replace zeros with NaN first (so therwon't affect the mean)
df.replace(0, pd.NA, inplace=True)

# Fill NaN with the mean of the column
df_filled = df.fillna(df.mean(numeric_only=True))

# Optional: Save to a new Excel file
df_filled.to_excel("data_filled.xlsx", index=False)

print("Missing values replaced with mean:")
print(df_filled)
```
